```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.layers import Rescaling

# Load and preprocess images using TensorFlow (if using it for ML)
train_dataset = image_dataset_from_directory(
    r"C:\Users\anura\.cache\kagglehub\datasets\tolgadincer\labeled-chest-xray-images\versions\1\chest_xray\train",  # Replace with the path where the images are stored
    image_size=(224, 224),  # Resize to 224x224 pixels for standard input size
    batch_size=32,
    label_mode='binary',  # 'binary' for 2 classes (NORMAL and PNEUMONIA)
    color_mode='rgb',  # Use RGB, change to 'grayscale' for grayscale images
)

# Optionally, normalize pixel values to [0, 1]
train_dataset = train_dataset.map(lambda x, y: (x / 255.0, y))

# If you want data augmentation:
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
])

train_dataset = train_dataset.map(lambda x, y: (data_augmentation(x), y))
```

```python
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2

# Define a simple CNN model or use a pre-trained model
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model.trainable = False  # Freeze the base model

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(1, activation='sigmoid')  # Binary classification output
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_dataset,
    epochs=10,  # You can change this based on your needs
)
```

```python
# Evaluate the model on the test dataset
test_dataset = image_dataset_from_directory(
    r"C:\Users\anura\.cache\kagglehub\datasets\tolgadincer\labeled-chest-xray-images\versions\1\chest_xray\test", 
    image_size=(224, 224), 
    batch_size=32,
    label_mode='binary',
    color_mode='rgb',
)

test_dataset = test_dataset.map(lambda x, y: (x / 255.0, y))  # Normalizing

test_loss, test_acc = model.evaluate(test_dataset)
print(f"Test Accuracy: {test_acc}")
```

```python
import numpy as np

# Load a sample image for prediction (replace with your image path)
img_path = r"C:\BACTERIA-1135262-0003.jpeg"  # Replace with the image path
img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

# Make a prediction
prediction = model.predict(img_array)
print("Prediction (0: NORMAL, 1: PNEUMONIA):", prediction)
```
